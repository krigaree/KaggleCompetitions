```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import keras
```

```python
dataset = pd.read_csv('train.csv')
dataset = dataset.dropna(subset=['Age', 'Embarked'])
#X = dataset.loc[:, ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].values
X = dataset.iloc[:, [2, 4, 5, 6, 7, 9, 11]].values
#X = dataset[]
y = dataset.iloc[:, 1].values
```

```python
# Swap NaN to mean
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
imputer = imputer.fit(X[:,2].reshape(-1,1))
X[:, 2] = imputer.transform(X[:, 2].reshape(-1,1)).reshape(1,-1)
#X
```

```python
# encoding sex
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
X[:,1] = class_le.fit_transform(X[:,1])
X[:,6] = class_le.fit_transform(X[:,6])
```

```python
# categorical
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(categorical_features=[0,1,3,4])
X = ohe.fit_transform(X).toarray()
```

```python
# Spliting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1042)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# X_train[:,[5,6,7,8]] = sc.fit_transform(X_train[:,[5,6,7,8]])
# X_test[:,[5,6,7,8]] = sc.transform(X_test[:,[5,6,7,8]])
X_train[:,[18,19]] = sc.fit_transform(X_train[:,[18,19]])
X_test[:,[18,19]] = sc.transform(X_test[:,[18,19]])
```

```python
m = X_train.shape[0]
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

cl = Sequential()

cl.add(Dense(4, activation='relu', input_dim=X_train.shape[1], kernel_initializer='uniform'))

#cl.add(Dropout(0.5))

cl.add(Dense(3, activation='relu', kernel_initializer='uniform'))

#cl.add(Dropout(0.8))

cl.add(Dense(3, activation='relu', kernel_initializer='uniform'))

#cl.add(Dropout(0.5))

cl.add(Dense(2, activation='relu', kernel_initializer='uniform'))

cl.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))

cl.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

cl.fit(X_train, y_train, batch_size=32, epochs=500, validation_split=0.0)
```

```python
y_pred = cl.predict(X_test)
y_pred_train = cl.predict(X_train)
y_pred = (y_pred > 0.5)
y_pred_train = (y_pred_train > 0.5)
```

```python
# CM
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
prec_test = (cm[0,0] + cm[1,1]) / sum(sum(cm))
prec_train = (cm2[0,0] + cm2[1,1]) / sum(sum(cm2))
print(cm)
```
